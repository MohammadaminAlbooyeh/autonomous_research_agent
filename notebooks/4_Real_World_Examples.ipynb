# Real-World Research Examples

This notebook demonstrates practical use cases of the autonomous research agent.

```python
import sys
sys.path.append('..')

from backend.agents.research_agent import ResearchAgent
from backend.services.research_service import ResearchService

agent = ResearchAgent()
service = ResearchService()
```

## Example 1: Technology Trend Analysis

Research the latest trends in a specific technology area.

```python
result = agent.execute("Edge Computing trends 2025", depth="shallow")
print(f"Topic: {result['topic']}")
print(f"Sources found: {len(result['sources'])}")
print(f"Report: {result['report']['title']}")
```

## Example 2: Using the API Service

Research tasks can be created via the service layer, which persists results to the database.

```python
task = service.create_task("Renewable Energy Storage", depth="shallow")
print(f"Task ID: {task.id}")
print(f"Status: {task.status}")
print(f"Sources: {len(task.sources)}")
print(f"Findings: {len(task.findings)}")
if task.report:
    print(f"Report sections: {len(task.report.sections)}")
```

## Example 3: Exporting Reports

Reports can be exported in Markdown or plain text format.

```python
from backend.services.report_service import ReportService

report_svc = ReportService()
if task.report:
    md = report_svc._to_markdown(task.report)
    print(md[:500] + "\n...")
```
