# Web Research with the Research Agent

This notebook demonstrates how the agent performs web searches and scrapes content.

```python
import sys
sys.path.append('..')

from backend.agents.web_searcher import WebSearcherAgent
from backend.tools.google_search import GoogleSearchTool
from backend.tools.web_scraper import WebScraperTool

searcher = WebSearcherAgent()
search_tool = GoogleSearchTool()
scraper = WebScraperTool()
```

## Generating Search Queries

The agent generates multiple search queries from a single topic.

```python
topic = "Quantum Computing"
queries = searcher.generate_queries(topic, depth="medium")
print("Generated queries:")
for q in queries:
    print(f"  - {q}")
```

## Running a Search

Search results include URLs, titles, and snippets.

```python
results = search_tool.search("Quantum computing latest breakthroughs", 5)
for i, r in enumerate(results, 1):
    print(f"{i}. {r['title']}")
    print(f"   URL: {r['url']}")
    print(f"   Snippet: {r['snippet'][:100]}...")
    print()
```
