# Autonomous Research Agent - Basics

This notebook introduces the core `ResearchAgent` and demonstrates how to use it.

```python
import sys
sys.path.append('..')

from backend.agents.research_agent import ResearchAgent

agent = ResearchAgent()
print("ResearchAgent initialized successfully")
```

## Running a Research Task

Execute a research task with a topic and depth level.

```python
result = agent.execute("Artificial Intelligence trends", depth="shallow")

print(f"Topic: {result['topic']}")
print(f"Queries: {result['queries']}")
print(f"Sources found: {len(result['sources'])}")
print(f"Findings: {len(result['findings'])}")
```

## Inspecting the Report

The report contains an executive summary, sections, and citations.

```python
report = result['report']
print(f"Title: {report['title']}")
print(f"\nExecutive Summary:\n{report['executive_summary']}")
print(f"\nSections: {len(report['sections'])}")
for s in report['sections']:
    print(f"  - {s['heading']}")
```
